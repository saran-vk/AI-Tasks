```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,0])
```

```python
m1 = MLPClassifier(
    hidden_layer_sizes=(),
    activation='logistic',
    max_iter=5000,
    random_state=42
)

m1.fit(X, y)

pred1 = m1.predict(X)
```

```python
print("WITHOUT HIDDEN LAYER")

for inp, pred in zip(X, pred1):
    print(inp, "->", pred)

acc1 = accuracy_score(y, pred1)

print("Accuracy =", acc1)
```

```python
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, 0.01),
    np.arange(y_min, y_max, 0.01)
)

Z = m1.predict(
    np.c_[xx.ravel(), yy.ravel()]
)

Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(6,5))
plt.contourf(xx, yy, Z, alpha=0.5)

plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    edgecolors='k',
    s=100
)

plt.title("Without Hidden Layer")
plt.show()
```

```python
m2 = MLPClassifier(
    hidden_layer_sizes=(4,),
    activation='logistic',
    max_iter=5000,
    random_state=42
)

m2.fit(X, y)
pred2 = m2.predict(X)
```

```python
print("\nWITH HIDDEN LAYER")

for inp, pred in zip(X, pred2):
    print(inp, "->", pred)

acc2 = accuracy_score(y, pred2)

print("Accuracy =", acc2)
```

```python
Z = m2.predict(
    np.c_[xx.ravel(), yy.ravel()]
)

Z = Z.reshape(xx.shape)

plt.figure(figsize=(6,5))
plt.contourf(xx, yy, Z, alpha=0.5)
plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    edgecolors='k',
    s=100
)

plt.title("With Hidden Layer (4 Neurons)")

plt.show()
```

```python
print("\nCOMPARISON")
print("Without Hidden Layer Accuracy =", acc1)
print("With Hidden Layer Accuracy =", acc2)

plt.figure(figsize=(6,4))
plt.bar(
    ["Without Hidden Layer",
     "With Hidden Layer"],
    [acc1, acc2]
)

plt.ylabel("Accuracy")
plt.title("Accuracy Comparison")
plt.show()
```
